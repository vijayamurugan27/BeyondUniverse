#### from these we are going to discuss , the file formats of JSON, CSV, xlsx

Text Files (.txt):

    Python's built-in open() function can be used to read and write text files.

```python
with open('file.txt', 'r') as f:
    content = f.read()
```

```python
print(content)
```

CSV Files (.csv):

    Use the csv module for reading and writing CSV files.

```python
import csv

with open('tips.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

JSON Files (.json):

    Use the json module for reading and writing JSON files.

```python
import json

with open('data.json', 'r') as f:
    data = json.load(f)
```

```python
print(data)
```

Excel Files (.xlsx, .xls):

    Use the openpyxl or xlrd (for older .xls format) libraries for reading and writing Excel files.
